```python
import ROOT
%jsroot on
from tdrStyle import *
setTDRStyle()
f = ROOT.TFile("ttHHadronicLoose_test_histograms2017.root")

h_max = {}
keys = ["Data", "DiPhoton", "GammaJets", "QCD", "TTGG", "TTGJets", "TTJets"]

for key_ in keys:
    
    if key_ == "DiPhoton":
        h_max[key_] = f.Get("hPhotonMaxIDMVA_"+key_+"GenPhoton_2")
    elif key_ == "GammaJets":
        h_max[key_] = f.Get("hPhotonMaxIDMVA_"+key_+"GenPhoton_1")
    elif key_ == "QCD":
        h_max[key_] = f.Get("hPhotonMaxIDMVA_"+key_+"GenPhoton_0")
    else:
    
        h_max[key_] = f.Get("hPhotonMaxIDMVA_"+key_)

h_max["DiPhoton"].Scale(0.22)
h_max["GammaJets"].Scale(7.39)
h_max["QCD"].Scale(2.57)

h_max_ttX = h_max["TTGG"].Clone("h_max_ttX")    
h_max_ttX.Add(h_max["TTGJets"])
h_max_ttX.Add(h_max["TTJets"])

stack = ROOT.THStack('stack','stack')                                                                                                                                                                        
stack.Add(h_max_ttX)
stack.Add(h_max["DiPhoton"])
stack.Add(h_max["GammaJets"])
stack.Add(h_max["QCD"])

total = h_max["DiPhoton"].Clone("total")
total.Add(h_max["GammaJets"])
total.Add(h_max["QCD"])
total.Add(h_max_ttX)

h_max["Data"].SetMarkerStyle(20)
h_max["DiPhoton"].SetFillColor(ROOT.TColor.GetColor("#41A90C"))
h_max["GammaJets"].SetFillColor(ROOT.TColor.GetColor("#FAAD12"))
h_max["QCD"].SetFillColor(ROOT.TColor.GetColor("#F98F6F"))
h_max_ttX.SetFillColor(ROOT.TColor.GetColor("#2C85D2"))              
              
c1 = ROOT.TCanvas("c1", "", 500, 800)
#c1.SetLogy()
pad1 = ROOT.TPad("pad1", "pad1", 0, 0.3, 1, 1.0)
pad1.SetBottomMargin(0)
#c1.SetGridy()                                                                                                                                                                                              
pad1.Draw()
pad1.cd()

dummy = ROOT.TH1D("dummy","dummy",1,-1,1)
dummy.SetMinimum(0.1)
dummy.SetMaximum(h_max["Data"].GetMaximum() * 1.5)
dummy.SetLineColor(0)
dummy.SetMarkerColor(0)
dummy.SetLineWidth(0)
dummy.SetMarkerSize(0)
dummy.GetYaxis().SetTitle("Events")
#dummy.GetXaxis().SetNdivisions(105)                                                                                                                                                                        
#dummy.GetYaxis().SetTitleOffset(1.3)                                                                                                                                                                       
dummy.GetXaxis().SetTitle("MaxIDMVA")
#dummy.GetXaxis().SetLabelOffset(99)                                                                                                                                                                        
dummy.SetTitle("")
dummy.Draw()
              
h_max["Data"].Draw("ep same")
stack.Draw("HIST same")
h_max["Data"].Draw("ep same")
total.SetFillColor(ROOT.TColor.GetColor("#8D8D84"))
total.SetFillStyle(3001)
total.Draw("pe2 same")

legend = ROOT.TLegend(0.35,0.65,0.75,0.9)
legend.AddEntry(h_max["Data"],"Data",'ep')
legend.AddEntry(h_max["QCD"],"QCD",'f')
legend.AddEntry(h_max["GammaJets"],"GammaJets",'f')
legend.AddEntry(h_max["DiPhoton"],"DiPhoton",'f')
legend.AddEntry(h_max_ttX,"ttX",'f')
legend.SetTextSize(0.03)
legend.SetLineWidth(1)
legend.SetFillColor(0)
legend.SetBorderSize(1)
legend.Draw("same")

c1.cd()
pad2 = ROOT.TPad("pad2", "pad2", 0, 0.05, 1, 0.3)
pad2.SetTopMargin(0)
pad2.SetBottomMargin(0.3)
pad2.SetGridy()
pad2.Draw()
pad2.cd()


#make ratio plot                                                                                                                                                                                            
ratio = h_max["Data"].Clone("ratio")
#ratio = TRatioPlot(stack, h_data)                                                                                                                                                                          
ratio.SetMinimum(0.001)                                                                                                                                                                                     
ratio.SetMaximum(1.999)
ratio.Sumw2()
ratio.SetStats(0)
ratio.Divide(total)
ratio.SetMarkerStyle(20)
ratio.Draw('e1p')

ratio.SetTitle("")
ratio.GetYaxis().SetTitle("Data/MC ratio")
ratio.GetXaxis().SetTitle("MaxIDMVA")
ratio.GetYaxis().SetTitleSize(20)
ratio.GetYaxis().SetTitleFont(43)
ratio.GetYaxis().SetTitleOffset(1.8)
ratio.GetYaxis().SetLabelFont(43)
ratio.GetYaxis().SetLabelSize(20)
ratio.GetXaxis().SetTitleSize(20)
ratio.GetXaxis().SetTitleFont(43)
ratio.GetXaxis().SetTitleOffset(4.)
ratio.GetXaxis().SetLabelFont(43)
ratio.GetXaxis().SetLabelSize(25)
ratio.GetYaxis().SetNdivisions(204)
c1.Draw()
c1.SaveAs("/home/users/hmei/public_html/2018/20181018_ttH_had/test_withWeight.png")
c1.SaveAs("/home/users/hmei/public_html/2018/20181018_ttH_had/test_withWeight.pdf")
```
